## Comparison of groups in demographic and clinical parameters at baseline
Comparison of control and experimental groups using t-test for selected parameters.

```python
from scipy.stats import fisher_exact, ttest_ind, mannwhitneyu
import numpy as np
import pandas as pd

# Baseline demographic characteristics table with SEHEPS subgroups
# Define mapping for display names and units
baseline_params = [
    ('gender', 'Sex (Male/Female)'),
    ('age', 'Age (y)'),
    ('height', 'Height (m)'),
    ('weight', 'Weight (kg)'),
    ('bmi', 'BMI'),
    ('education_years', 'Education (y)'),
    ('duration', 'Disease duration (y)'),
    ('w0_hy', 'Hoehn & Yahr'),
    ('w0_updrsiii', 'MDS-UPDRS III, ON'),
    ('moca', 'MoCa'),
    ('led_mg', 'LEDD (mg/day)'),
    ('seheps', 'SEHEPS'),
    ('w0_mep', 'Maximum expiratory pressure (cmH2O)')
]

rows = []
for col, label in baseline_params:
    if col == 'gender':
        # Count males and females for all and SEHEPS subgroups
        def count_mf(df):
            return (df['gender'] == 'M').sum(), (df['gender'] == 'F').sum()
        
        cg_m, cg_f = count_mf(CG)
        eg_m, eg_f = count_mf(EG)
        cg_m1, cg_f1 = count_mf(CG[CG['w8_seheps'] < 59])
        cg_m2, cg_f2 = count_mf(CG[CG['w8_seheps'] >= 59])
        eg_m1, eg_f1 = count_mf(EG[EG['w8_seheps'] < 59])
        eg_m2, eg_f2 = count_mf(EG[EG['w8_seheps'] >= 59])
        
        
        cg_val = f"{cg_m}/{cg_f} [ {cg_m1}/{cg_f1}; {cg_m2}/{cg_f2} ]"
        eg_val = f"{eg_m}/{eg_f} [ {eg_m1}/{eg_f1}; {eg_m2}/{eg_f2} ]"
        # Fisher's exact test for gender distribution (all)
        table = np.array([[cg_m, cg_f], [eg_m, eg_f]])
        _, p_val_gender = fisher_exact(table)
        # Fisher's exact test for SEHEPS subgroups in CG and EG
        table_cg = np.array([[cg_m1, cg_f1], [cg_m2, cg_f2]]) if (cg_m1+cg_f1>0 and cg_m2+cg_f2>0) else np.nan
        table_eg = np.array([[eg_m1, eg_f1], [eg_m2, eg_f2]]) if (eg_m1+eg_f1>0 and eg_m2+eg_f2>0) else np.nan
        pval_cg = fisher_exact(table_cg)[1] if isinstance(table_cg, np.ndarray) else np.nan
        pval_eg = fisher_exact(table_eg)[1] if isinstance(table_eg, np.ndarray) else np.nan
        row_format = [label, cg_val, eg_val, f"{p_val_gender:.3f} [{pval_cg:.3f};{pval_eg:.3f}]"]
        rows.append(row_format)
        continue
    
    
    # All values
    cg_vals = CG[col].dropna()
    eg_vals = EG[col].dropna()
    # SEHEPS subgroups
    cg1 = CG[CG['w8_seheps'] < 59][col].dropna()
    cg2 = CG[CG['w8_seheps'] >= 59][col].dropna()
    eg1 = EG[EG['w8_seheps'] < 59][col].dropna()
    eg2 = EG[EG['w8_seheps'] >= 59][col].dropna()
    # Means and SDs
    cg_mean, cg_std = cg_vals.mean(), cg_vals.std()
    eg_mean, eg_std = eg_vals.mean(), eg_vals.std()
    cg1_mean, cg1_std = cg1.mean(), cg1.std() if len(cg1)>0 else (np.nan, np.nan)
    cg2_mean, cg2_std = cg2.mean(), cg2.std() if len(cg2)>0 else (np.nan, np.nan)
    eg1_mean, eg1_std = eg1.mean(), eg1.std() if len(eg1)>0 else (np.nan, np.nan)
    eg2_mean, eg2_std = eg2.mean(), eg2.std() if len(eg2)>0 else (np.nan, np.nan)
    # Format: main [<59; â‰¥59]
    cg_val = f"{cg_mean:.1f} ({cg_std:.1f}) [ {cg1_mean:.1f} ({cg1_std:.1f}); {cg2_mean:.1f} ({cg2_std:.1f}) ]"
    eg_val = f"{eg_mean:.1f} ({eg_std:.1f}) [ {eg1_mean:.1f} ({eg1_std:.1f}); {eg2_mean:.1f} ({eg2_std:.1f}) ]"
    # Test for normality in subgroups
    cg1_norm = stats.kstest((cg1 - cg1.mean()) / cg1.std(ddof=0), 'norm').pvalue > 0.05 if len(cg1)>2 else False
    cg2_norm = stats.kstest((cg2 - cg2.mean()) / cg2.std(ddof=0), 'norm').pvalue > 0.05 if len(cg2)>2 else False
    eg1_norm = stats.kstest((eg1 - eg1.mean()) / eg1.std(ddof=0), 'norm').pvalue > 0.05 if len(eg1)>2 else False
    eg2_norm = stats.kstest((eg2 - eg2.mean()) / eg2.std(ddof=0), 'norm').pvalue > 0.05 if len(eg2)>2 else False
    # p-value for difference between SEHEPS subgroups in CG and EG
    if cg1_norm and eg1_norm and len(cg1)>1 and len(eg1)>1:
        _, pval_cg = ttest_ind(cg1, eg1, nan_policy='omit')
    else:
        try:
            _, pval_cg = mannwhitneyu(cg1, eg1, alternative='two-sided')
        except:
            pval_cg = np.nan
    if cg2_norm and eg2_norm and len(cg2)>1 and len(eg2)>1:
        _, pval_eg = ttest_ind(cg2, eg2, nan_policy='omit')
    else:
        try:
            _, pval_eg = mannwhitneyu(cg2, eg2, alternative='two-sided')
        except:
            pval_eg = np.nan
            
            
    # p-value for CG vs EG (all)
    cg_norm = stats.kstest((cg_vals - cg_vals.mean()) / cg_vals.std(ddof=0), 'norm').pvalue > 0.05 if len(cg_vals)>2 else False
    eg_norm = stats.kstest((eg_vals - eg_vals.mean()) / eg_vals.std(ddof=0), 'norm').pvalue > 0.05 if len(eg_vals)>2 else False
    if cg_norm and eg_norm and len(cg_vals)>1 and len(eg_vals)>1:
        _, p_val = ttest_ind(cg_vals, eg_vals, nan_policy='omit')
        normality = '*'
    else:
        try:
            _, p_val = mannwhitneyu(cg_vals, eg_vals, alternative='two-sided')
        except:
            p_val = np.nan
        normality = ''
    row_format = [
        label,
        cg_val,
        eg_val,
        f"{p_val:.3f} [{pval_cg:.3f};{pval_eg:.3f}]{normality}"
        
    ]
    rows.append(row_format)

# Create DataFrame for display
baseline_table = pd.DataFrame(
    rows,
    columns=['Measure', 'Control group (n=38) [WEEK 8 SEHEPS <59;â‰¥59]', 'Experimental group (n=37) [WEEK 8 SEHEPS <59;â‰¥59]', 'CG vs EG P-value [WEEK 8 SEHEPS <59;â‰¥59]']    
)
display(baseline_table)
```

### Linear Mixed Effects Model (LMM) for adherence 
**Linear Mixed Effects Model (LMM):**
A linear mixed effects model was used to analyze adherence across time intervals, groups, and SEHEPS subgroups, accounting for repeated measures within subjects. The model formula was:
$$\text{adherence} \sim \text{group} \times \text{interval} \times \text{SEHEPS group} + (1|\text{ID})$$
where group (CG/EG), interval (W0-8/W8-24), and SEHEPS group (<59/â‰¥59) were included as fixed effects and their interactions, and subject ID was included as a random intercept to account for within-subject correlation. Model assumptions were checked using residual diagnostics (Shapiro-Wilk and Breusch-Pagan tests).
The model was fitted using the `statsmodels` Python package (function `mixedlm`).

```python

import pandas as pd
import statsmodels.formula.api as smf

# Prepare data for LMM: long format for repeated measures
adherence_long = pd.melt(
    df,
    id_vars=['ID', 'group', 'w8_seheps'],
    value_vars=['w0-8_total', 'w8-24_total'],
    var_name='interval',
    value_name='adherence'
)

# Create SEHEPS group (<59 vs >=59 at week 8)
adherence_long['seheps_group'] = (adherence_long['w8_seheps'] >= 59).astype(int)

# Categorical variables for model
adherence_long['interval'] = adherence_long['interval'].map({'w0-8_total': 'W0-8', 'w8-24_total': 'W8-24'})
adherence_long['group'] = adherence_long['group'].map({0: 'CG', 1: 'EG'})
adherence_long['seheps_group'] = adherence_long['seheps_group'].map({0: '<59', 1: 'â‰¥59'})

# Remove missing adherence values
adherence_long = adherence_long.dropna(subset=['adherence'])

# Fit the LMM: adherence ~ group * interval * seheps_group + (1|ID)
model = smf.mixedlm(
    'adherence ~ group * interval * seheps_group',
    data=adherence_long,
    groups=adherence_long['ID']
)
result = model.fit(reml=False)
print(result.summary())

```

**Main findings from the model:**

The estimated mean adherence in the reference group (Control Group, interval W0-8, SEHEPS <59). This value is statistically significant.

group[T.EG] (50.3, p=0.70):
The difference in adherence between the Experimental Group (EG) and Control Group (CG) in the W0-8 interval and SEHEPS <59. This difference is not statistically significant.

Adherence in the W8-24 interval is significantly lower than in W0-8 (for CG, SEHEPS <59).

The difference in adherence between SEHEPS â‰¥59 and <59 (for CG, W0-8) is not significant.

group[T.EG]:seheps_group[T.â‰¥59] (-15.9, p=0.94):
The interaction between group and SEHEPS is not significant.

interval[T.W8-24]:seheps_group[T.â‰¥59] (567.1, p=0.006):
The difference in adherence between intervals (W8-24 vs W0-8) is significantly larger for patients with SEHEPS â‰¥59.

group[T.EG]:interval[T.W8-24]:seheps_group[T.â‰¥59] (-426.5, p=0.14):
The three-way interaction is not statistically significant.

Group Var (7809.1):
This is the estimated variance in adherence between individual subjects (ID), after accounting for the fixed effects.
Summary:

Adherence generally decreases in the later interval (W8-24), but this decrease is less pronounced in the Experimental Group.

Other effects and interactions are not statistically significant.

**Conclusion:**
The intervention (EG) helps maintain higher adherence in the later phase, especially compared to the control group, and the effect of time interval is more pronounced in patients with higher SEHEPS scores.

```python
# --- Diagnostics: Check LMM assumptions (normality, homoscedasticity of residuals) ---
import matplotlib.pyplot as plt
import scipy.stats as stats
from statsmodels.stats.diagnostic import het_breuschpagan

# Get residuals
resid = result.resid
fitted = result.fittedvalues

# 1. QQ plot for normality
plt.figure(figsize=(5, 4))
stats.probplot(resid, dist="norm", plot=plt)
plt.title("QQ plot of LMM residuals")
plt.show()

# 2. Histogram of residuals
plt.figure(figsize=(5, 4))
plt.hist(resid, bins=20, edgecolor='k', alpha=0.7)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Histogram of LMM residuals")
plt.show()

# 3. Shapiro-Wilk test for normality
shapiro_stat, shapiro_p = stats.shapiro(resid)
print(f"Shapiro-Wilk test: W={shapiro_stat:.3f}, p={shapiro_p:.3g}")

# 4. Plot residuals vs fitted values (check for homoscedasticity)
plt.figure(figsize=(5, 4))
plt.scatter(fitted, resid, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted values")
plt.show()

# 5. Breusch-Pagan test for heteroscedasticity
bp_test = het_breuschpagan(resid, result.model.exog)
print(f"Breusch-Pagan test: LM stat={bp_test[0]:.2f}, p={bp_test[1]:.3g}")
```

**Diagnostics summary:**

- **Shapiro-Wilk test:** p < 0.001 â†’ The residuals are not normally distributed.
- **Breusch-Pagan test:** p = 0.016 â†’ There is evidence of heteroscedasticity (the variance of the residuals is not constant).

**Impact:** The model results should be interpreted with caution, as the assumptions of normality and homoscedasticity of residuals are not fully met. Robust estimates or data transformation may be considered, or the results can be supplemented with a sensitivity analysis.

```python
from scipy.stats import mannwhitneyu
import numpy as np
import pandas as pd
from stats_helpers import mean_ci, cohen_d

# Helper for row formatting
def row_format1(group, n, mean, std, ci, pval=None, d=None, dci=None):
    r = [group, n, f"{mean:.0f}", f"{std:.1f}", f"[{ci[0]:.0f}, {ci[1]:.0f}]"]
    if pval is not None:
        r.append(f"{pval:.2f}")
    else:
        r.append("")
    if d is not None:
        r.append(f"{d:.2f}")
        r.append(f"[{dci[0]:.2f}, {dci[1]:.2f}]")
    else:
        r.append("")
        r.append("")
    return r


table_oa = []
columns = ["Time interval", "Group", "n", "Mean", "SD", "95% CI", "P-value", "Cohen d", "Cohen d 95% CI" ]

# W0-8
cg0 = CG['w0-8_total'].dropna()
eg0 = EG['w0-8_total'].dropna()
mean_cg0, std_cg0, n_cg0, ci_cg0 = mean_ci(cg0)
mean_eg0, std_eg0, n_eg0, ci_eg0 = mean_ci(eg0)
stat0, p0 = mannwhitneyu(cg0, eg0, alternative='two-sided')
d0, d0ci = cohen_d(cg0, eg0)
table_oa.append(["W0-8", "CG"] + row_format1("CG", n_cg0, mean_cg0, std_cg0, ci_cg0, p0, d0, d0ci)[1:])
table_oa.append(["", "EG"] + row_format1("EG", n_eg0, mean_eg0, std_eg0, ci_eg0)[1:])

# W8-24
cg1 = CG['w8-24_total'].dropna()
eg1 = EG['w8-24_total'].dropna()
mean_cg1, std_cg1, n_cg1, ci_cg1 = mean_ci(cg1)
mean_eg1, std_eg1, n_eg1, ci_eg1 = mean_ci(eg1)
stat1, p1 = mannwhitneyu(cg1, eg1, alternative='two-sided')
d1, d1ci = cohen_d(cg1, eg1)
table_oa.append(["W8-24", "CG"] + row_format1("CG", n_cg1, mean_cg1, std_cg1, ci_cg1, p1, d1, d1ci)[1:])
table_oa.append(["", "EG"] + row_format1("EG", n_eg1, mean_eg1, std_eg1, ci_eg1)[1:])

adherence_table = pd.DataFrame(table_oa, columns=columns)
display(adherence_table)

```

### Overall Changes in MEP and Subgroup Analysis Based on SEHEPS at Weekly Intervals


```python

import numpy as np
import pandas as pd
from scipy.stats import  ttest_ind
from stats_helpers import mean_ci, cohen_d
from scipy.stats import ttest_rel

rows = []
def row_format2(
    interval, group, stats,g_pval=None, g_d=None, g_d_ci=None):
    row = [interval, group, stats['n'],
           f"{stats['mean_pre']:.1f} ({stats['std_pre']:.1f})", f"{stats['mean_post']:.1f} ({stats['std_post']:.1f})",
           f"{stats['mean_diff']:+.1f} ({stats['std_diff']:.1f})", f"[{stats['ci_diff'][0]:.1f}, {stats['ci_diff'][1]:.1f}]",
           f"{stats['pval']:.3f}",
           f"{stats['d']:.2f} [{stats['d_ci'][0]:.2f}, {stats['d_ci'][1]:.2f}]"
    ]
    if g_pval is not None:
        row.append(f"{g_pval:.3f}")
    else:
        row.append("")
    if g_d is not None and g_d_ci is not None:
        row.append(f"{g_d:.2f} [{g_d_ci[0]:.2f}, {g_d_ci[1]:.2f}]")
    else:
        row.append("")
    return row

# Helper to get stats from precomputed columns
def stats_from_diff(pre, post, diff):
    # Only keep entries where diff is not nan
    mask = ~diff.isna()
    pre = pre[mask]
    post = post[mask]
    diff = diff[mask]
    n = pre.count()
    mean_pre = pre.mean()
    std_pre = pre.std()
    mean_post = post.mean()
    std_post = post.std()
    mean_diff = diff.mean()
    std_diff = diff.std()
    ci_diff = mean_ci(diff)[3]
    # PÃ¡rovÃ½ t-test pre vs post
    pval = ttest_rel(pre, post, nan_policy='omit')[1]
    # Cohen's d pro pÃ¡rovÃ½ rozdÃ­l
    d = (mean_post - mean_pre) / np.sqrt((std_pre ** 2 + std_post ** 2) / 2)
    # CI pro Cohen's d (pÅ™ibliÅ¾nÄ›, 95%)
    se_d = np.sqrt((1/n) + (d**2/(2*n)))
    d_ci = (d - 1.96 * se_d, d + 1.96 * se_d)
    return {
        'n': n,
        'mean_pre': mean_pre,
        'std_pre': std_pre,
        'mean_post': mean_post,
        'std_post': std_post,
        'mean_diff': mean_diff,
        'std_diff': std_diff,
        'ci_diff': ci_diff,
        'pval': pval,
        'd': d,
        'd_ci': d_ci
    }

# Overall changes in MEP
# W0-W8
cg_stats = stats_from_diff(CG['w0_mep'], CG['w8_mep'], CG['w8-w0_mep'])
eg_stats = stats_from_diff(EG['w0_mep'], EG['w8_mep'], EG['w8-w0_mep'])



d, d_ci = cohen_d(CG['w8-w0_mep'], EG['w8-w0_mep'])
pval_between = ttest_ind(CG['w8-w0_mep'], EG['w8-w0_mep'], nan_policy='omit')[1]
rows.append(row_format2('W0-W8', 'CG', cg_stats, pval_between, d, d_ci))
rows.append(row_format2('', 'EG', eg_stats))


# W8-W24
cg_stats = stats_from_diff(CG['w8_mep'], CG['w24_mep'], CG['w24-w8_mep'])
eg_stats = stats_from_diff(EG['w8_mep'], EG['w24_mep'], EG['w24-w8_mep'])

d, d_ci = cohen_d(CG['w24-w8_mep'], EG['w24-w8_mep'])
pval_between = ttest_ind(CG['w24-w8_mep'], EG['w24-w8_mep'], nan_policy='omit')[1]
rows.append(row_format2('W8-W24', 'CG', cg_stats, pval_between, d, d_ci))
rows.append(row_format2('', 'EG', eg_stats))


# W0-W24

cg_stats = stats_from_diff(CG['w0_mep'], CG['w24_mep'], CG['w24-w0_mep'])
eg_stats = stats_from_diff(EG['w0_mep'], EG['w24_mep'], EG['w24-w0_mep'])

d, d_ci = cohen_d(CG['w24-w0_mep'], EG['w24-w0_mep'])
pval_between = ttest_ind(CG['w24-w0_mep'], EG['w24-w0_mep'], nan_policy='omit')[1]
rows.append(row_format2('W0-W24', 'CG', cg_stats, pval_between, d, d_ci))
rows.append(row_format2('', 'EG', eg_stats))


# Subgroup Analysis Based on SEHEPS at Week 8
# W8-W24, SEHEPS <59
cg_sub = CG[CG['w8_seheps'] < 59]
eg_sub = EG[EG['w8_seheps'] < 59]
cg_stats = stats_from_diff(cg_sub['w8_mep'], cg_sub['w24_mep'], cg_sub['w24-w8_mep'])
eg_stats = stats_from_diff(eg_sub['w8_mep'], eg_sub['w24_mep'], eg_sub['w24-w8_mep'])

d, d_ci = cohen_d(cg_sub['w24-w8_mep'], eg_sub['w24-w8_mep'])
pval_between = ttest_ind(cg_sub['w24-w8_mep'], eg_sub['w24-w8_mep'], nan_policy='omit')[1]
rows.append(row_format2('W8-W24', 'CG SEHEPS <59', cg_stats, pval_between, d, d_ci))
rows.append(row_format2('', 'EG SEHEPS <59', eg_stats))


# W8-W24, SEHEPS >=59
cg_sub = CG[CG['w8_seheps'] >= 59]
eg_sub = EG[EG['w8_seheps'] >= 59]
cg_stats = stats_from_diff(cg_sub['w8_mep'], cg_sub['w24_mep'], cg_sub['w24-w8_mep'])
eg_stats = stats_from_diff(eg_sub['w8_mep'], eg_sub['w24_mep'], eg_sub['w24-w8_mep'])
pval_between = ttest_ind(cg_sub['w24-w8_mep'], eg_sub['w24-w8_mep'], nan_policy='omit')[1]
d, d_ci = cohen_d(cg_sub['w24-w8_mep'], eg_sub['w24-w8_mep'])
rows.append(row_format2('W8-W24', 'CG SEHEPS â‰¥59', cg_stats, pval_between, d, d_ci))
rows.append(row_format2('', 'EG SEHEPS â‰¥59', eg_stats))

columns = ['Interval', 'Group', 'n', 'Pre- Mean (SD)', 'Post- Mean (SD)', 'Mean (SD) Î”', '95% CI Î”', 'P-value  Î”','Cohen d (95% CI) Î”','P-value EG vs CG', 'Cohen d (95% CI) EG vs CG']
table3 = pd.DataFrame(rows, columns=columns)
display(table3)
```

**Main results for MEP changes**
